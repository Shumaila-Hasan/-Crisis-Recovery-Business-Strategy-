```python
def clean_prep_time(x):
    if pd.isna(x):
        return None
    
    x = str(x).strip()
    
    
    if '-' in x:
        a, b = x.split('-')
        return (int(a) + int(b)) / 2
    
    
    elif '<=' in x:
        return int(x.replace('<=', ''))
    
    
    elif '>' in x:
        return int(x.replace('>', ''))
    
    
    elif '+' in x:
        return int(x.replace('+', ''))
    
    
    else:
        return int(x)

# Apply
restaurant['avg_prep_time_min'] = restaurant['avg_prep_time_min'].apply(clean_prep_time)
```

```python
query = '''

select  DATE_TRUNC('month' , order_timestamp) as month,
count(*) as total_orders
from orders
WHERE is_cancelled = 'N'
group by month
order by month asc
'''
orders= pd.read_sql(query, engine)
orders



```

# City-wise Decline %

```python
query = '''
WITH pre AS (
    SELECT r.city, COUNT(o.order_id) AS pre_orders
    FROM orders o
    JOIN restaurant r ON o.restaurant_id = r.restaurant_id
    WHERE o.order_timestamp BETWEEN '2025-01-01' AND '2025-05-31'
      AND is_cancelled = 'N'
    GROUP BY r.city
),
crisis AS (
    SELECT r.city, COUNT(o.order_id) AS crisis_orders
    FROM orders o
    JOIN restaurant r ON o.restaurant_id = r.restaurant_id
    WHERE o.order_timestamp BETWEEN '2025-06-01' AND '2025-09-30'
      AND is_cancelled = 'N'
    GROUP BY r.city
)
SELECT 
    p.city,
    pre_orders,
    crisis_orders,
    ROUND((pre_orders - crisis_orders)*100.0/pre_orders,2) AS decline_pct
FROM pre p
JOIN crisis c ON p.city = c.city
ORDER BY decline_pct DESC
LIMIT 5
'''
order_trend= pd.read_sql(query, engine)
order_trend
```

```python
import matplotlib.pyplot as plt
order_trend.set_index('city')[['pre_orders', 'crisis_orders']].plot(
    kind='bar',
    figsize=(6,4),
    title='Pre vs Crisis Orders by City'
)

plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Top Restaurants Decline

```python
query = '''
WITH pre AS (
    SELECT restaurant_id, COUNT(*) AS pre_orders
    FROM orders
    WHERE order_timestamp BETWEEN '2025-01-01' AND '2025-05-31'
      AND is_cancelled = 'N'
    GROUP BY restaurant_id
    
),
crisis AS (
    SELECT restaurant_id, COUNT(*) AS crisis_orders
    FROM orders
    WHERE order_timestamp BETWEEN '2025-06-01' AND '2025-09-30'
      AND is_cancelled = 'N'
    GROUP BY restaurant_id
)
SELECT 
    r.restaurant_name,
    pre_orders,
    crisis_orders,
    ROUND((pre_orders - crisis_orders)*100.0/pre_orders,2) AS decline_pct
FROM pre p
JOIN crisis c ON p.restaurant_id = c.restaurant_id
JOIN restaurant r ON r.restaurant_id = p.restaurant_id
ORDER BY decline_pct DESC
LIMIT 10;

'''

restaurants_decline = pd.read_sql(query,engine)
restaurants_decline
```

# Delivery SLA

```python
query = '''
SELECT 
    DATE_TRUNC('month', o.order_timestamp) AS month,
    AVG(actual_delivery_time_mins) AS avg_actual,
    AVG(expected_delivery_time_mins) AS avg_expected,
    AVG(actual_delivery_time_mins - expected_delivery_time_mins) AS delay
FROM delivery_performance d
JOIN orders o ON o.order_id = d.order_id
GROUP BY 1
ORDER BY 1;
'''
sla = pd.read_sql(query,engine)
sla
```

# Ratings Trend

```python
query = '''
SELECT 
    DATE_TRUNC('month', review_timestamp::timestamp) AS month,
    AVG(rating) AS avg_rating
FROM ratings
WHERE review_timestamp IS NOT NULL
GROUP BY 1
ORDER BY 1;

'''
rat_trend = pd.read_sql(query,engine)
rat_trend 
```

# Revenue Impact

```python
query = '''
SELECT 
    CASE 
      WHEN order_timestamp BETWEEN '2025-01-01' AND '2025-05-31' THEN 'Pre'
      ELSE 'Crisis'
    END AS phase,
    SUM(total_amount) AS revenue
FROM orders
WHERE is_cancelled = 'N'
GROUP BY phase;

'''
revenue_impact =  pd.read_sql(query,engine)
revenue_impact 
```
